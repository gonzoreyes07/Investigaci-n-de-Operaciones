```python
import networkx as nx
import matplotlib.pyplot as plt

# Create the graph
G = nx.Graph()
G.add_nodes_from(["Inicio", "A", "B", "C", "D", "E", "F", "G", "Fin"])
G.add_edge('Inicio', 'A')
G.add_edge('Inicio', 'B')
G.add_edge('Inicio', 'C')
G.add_edge('A', 'D')
G.add_edge('B', 'E')
G.add_edge('C', 'F')
G.add_edge('D', 'G')
G.add_edge('E', 'G')
G.add_edge('G', 'Fin')
G.add_edge('F', 'Fin')

# Define the layout
pos = nx.spring_layout(G, seed=42)

# Draw the graph
plt.figure(figsize=(5, 5))

# Draw all nodes with default color
nx.draw_networkx_nodes(G, pos, node_color='greenyellow', node_size=600)

# Highlight nodes in the paths A -> D -> G (red) and B -> E -> G (blue)
highlight_nodes_red = ['A', 'D', 'G']
highlight_nodes_blue = ['B', 'E', 'G']
nx.draw_networkx_nodes(G, pos, nodelist=highlight_nodes_red, node_color='aquamarine', node_size=600)
nx.draw_networkx_nodes(G, pos, nodelist=highlight_nodes_blue, node_color='violet', node_size=600)

# Draw all edges with default color (gray)
nx.draw_networkx_edges(G, pos, edge_color='black')

# Highlight edges in the paths A -> D -> G (red) and B -> E -> G (blue)
highlight_edges_red = [('A', 'D'), ('D', 'G')]
highlight_edges_blue = [('B', 'E'), ('E', 'G')]
nx.draw_networkx_edges(G, pos, edgelist=highlight_edges_red, edge_color='aquamarine', width=2)
nx.draw_networkx_edges(G, pos, edgelist=highlight_edges_blue, edge_color='violet', width=2)

# Draw labels
nx.draw_networkx_labels(G, pos, font_weight='bold')
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

# Set title
plt.title("Rutas críticas")
plt.show()
```
