  # **PROBABILIDADES DEL TABLERO DE COLORES**
  
  Planteamiento:
  
  Se modela la evolución de las probabilidades \(x_n\), \(y_n\), \(z_n\), \(t_n\) usando la matriz de transición \(A\). El objetivo es determinar si, a largo plazo, todas las probabilidades tienden a \(0.25\), indicando que todos los sectores son igual de probables.
    
  Matriz de transición \(A\):
  $$
  A = \begin{pmatrix}
 \frac{1}{4} & 0 & \frac{1}{4} & \frac{2}{4} \\\\
  \frac{2}{4} & \frac{1}{4} & 0 & \frac{1}{4} \\\\
  \frac{1}{4} & \frac{2}{4} & \frac{1}{4} & 0 \\\\
  0 & \frac{1}{4} & \frac{2}{4} & \frac{1}{4}
  \end{pmatrix}
  $$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Definimos la matriz de transición A
A = np.array([[1/4, 0, 1/4, 2/4],
                [2/4, 1/4, 0, 1/4],
                [1/4, 2/4, 1/4, 0],
                [0, 1/4, 2/4, 1/4]])
```

```python
# Condiciones iniciales Comenazamos con probabilidad 1 en x y 0 en las otras
X0 = np.array([1.0, 0.0, 0.0, 0.0])  # Vector inicial

# Número de iteraciones
n = 50

# Listas para almacenar las probabilidades
x_n = [X0[0]]
y_n = [X0[1]]
z_n = [X0[2]]
t_n = [X0[3]]
```

```python
# Iteramos para calcular las probabilidades
x_actual = X0
for i in range(n):
    x_nuevo = A @ x_actual  # Multiplicamos matricialmente
    x_n.append(x_nuevo[0])
    y_n.append(x_nuevo[1])
    z_n.append(x_nuevo[2])
    t_n.append(x_nuevo[3])
    x_actual = x_nuevo  # Actualizamos el estado actual
```

```python
# Graficamos las probabilidades
plt.figure(figsize=(10, 6))
plt.plot(x_n, label='x_n')
plt.plot(y_n, label='y_n')
plt.plot(z_n, label='z_n')
plt.plot(t_n, label='t_n')

plt.xlabel('Iteración (n)')
plt.ylabel('Probabilidad')
plt.title('Evolución de las Probabilidades con el Tiempo')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Imprimir las probabilidades finales
print("Probabilidades finales después de", n, "iteraciones:")
print("xₙ =", x_n[-1])
print("yₙ =", y_n[-1])
print("zₙ =", z_n[-1])
print("tₙ =", t_n[-1])
```

  ### Análisis
Si las probabilidades convergen aproximadamente a 0.25, todos los sectores son igual de probables.
