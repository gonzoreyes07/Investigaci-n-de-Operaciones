  ### Planteamiento
  Se modela la evolución de las probabilidades \(x_n\), \(y_n\), \(z_n\), \(t_n\) usando la matriz de transición \(A\). El objetivo es determinar si, a largo plazo, todas las probabilidades tienden a \(0.25\), indicando que todos los sectores son igual de probables.
    
  Matriz de transición \(A\):
  $$
  A = \begin{pmatrix}
 \frac{1}{4} & 0 & \frac{1}{4} & \frac{2}{4} \\\\
  \frac{2}{4} & \frac{1}{4} & 0 & \frac{1}{4} \\\\
  \frac{1}{4} & \frac{2}{4} & \frac{1}{4} & 0 \\\\
  0 & \frac{1}{4} & \frac{2}{4} & \frac{1}{4}
  \end{pmatrix}
  $$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Definimos la matriz de transición A
A = np.array([[1/4, 0, 1/4, 2/4],
                [2/4, 1/4, 0, 1/4],
                [1/4, 2/4, 1/4, 0],
                [0, 1/4, 2/4, 1/4]])
```

```python
# Número de iteraciones
n_steps = 2

# Vector inicial de probabilidades (ejemplo: [1, 0, 0, 0])
X = np.array([1.0, 0.0, 0.0, 0.0])

# Listas para almacenar la evolución de las probabilidades
x_n = [X[0]]
y_n = [X[1]]
z_n = [X[2]]
t_n = [X[3]]
```

```python
# Iterar para calcular las probabilidades
for _ in range(n_steps - 1):
  X = A @ X  # Multiplicación matricial
  x_n.append(X[0])
  y_n.append(X[1])
  z_n.append(X[2])
  t_n.append(X[3])
```

```python
# Verificar que las probabilidades sumen 1 (normalización)
print("Suma de probabilidades en cada paso:", [sum(X) for X in zip(x_n, y_n, z_n, t_n)])

# Graficar las probabilidades
plt.figure(figsize=(10, 6))
plt.plot(x_n, label='x_n')
plt.plot(y_n, label='y_n')
plt.plot(z_n, label='z_n')
plt.plot(t_n, label='t_n')
plt.axhline(y=0.25, color='r', linestyle='--', label='Equilibrio (0.25)')
plt.xlabel('Número de iteraciones')
plt.ylabel('Probabilidad')
plt.title('Evolución de las Probabilidades con el Tiempo')
plt.legend()
plt.grid(True)
plt.show()
```

  ### Análisis
  La gráfica muestra cómo las probabilidades \(x_n\), \(y_n\), \(z_n\), \(t_n\) evolucionan con el tiempo. Si todas convergen a \(0.25\), indica que los sectores son equally probables a largo plazo. Observar la tendencia en las últimas iteraciones.
